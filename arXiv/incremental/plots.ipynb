```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets = ['led', 'heart', 'zoo', 'votes']
for dataset in datasets:
    
    accs = pickle.load(open("vfdt_test_utility_"+dataset+".pkl", "rb" ))
    plt.figure()
    plt.plot(np.mean(accs, axis=0))
    plt.fill_between(range(accs.shape[1]),np.mean(accs, axis=0)-np.std(accs, axis=0), np.mean(accs, axis=0)+np.std(accs, axis=0),alpha=0.2)
    plt.savefig('Test_utility_'+dataset+'.pdf', format='pdf')
```
